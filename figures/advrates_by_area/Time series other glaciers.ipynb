```python
ti = np.arange(date2num(d.time[0]), date2num(d.time[-1])+1, 0.5)
fs = 15 


for ewkey in ewkeys:
    figdir_ew = figdir + '%s/'%ewkey

    glkeys = d.glacier_sorting[ewkey]
    
    for glkey in glkeys:
        
        fig, ax = plt.subplots(figsize = (6, 3))
        
        advrate = d[glkey]

        advrate_pos = ip.interp1(date2num(d.time), advrate.copy(), ti)
        advrate_neg = ip.interp1(date2num(d.time), advrate.copy(), ti)
        advrate_pos[advrate_pos<0] = np.nan
        advrate_neg[advrate_neg>0] = np.nan
        
        fillkws = {'alpha':0.7, 'clip_on':False}

        ax.fill_between(ti, advrate_pos,  color = col.col('b'), **fillkws)
        ax.fill_between(ti, advrate_neg,  color = col.col('r'), **fillkws)
        ax.plot(d.time, advrate, '-', color = col.col('k'), lw = 1, clip_on = False)
        
        ax.invert_yaxis()
        ax.set_ylabel('m day$^{-1}$', fontsize = fs)
        
        ax.set_title(glkey)
        pretty.lrbars(ax)
        ax.set_ylim(6, -14)

        pretty.axthick(ax)
        for yr in np.arange(2017, 2022):
            vl_kws = {'alpha':0.3}
            pl.vline(dt.date(yr, 10, 1), ax = ax, **vl_kws)
            pl.vline(dt.date(yr, 4, 1), ax = ax, **vl_kws, ls = ':')
            
            if yr//2!=yr/2:
                pl.fill_xrange(dt.date(yr, 1, 1), dt.date(yr, 12, 31), 
                           ax, color = col.col('g'), alpha = 0.2)

        pretty.rotate_y(ax)
        ax.set_xlim(dt.date(2016, 10, 1), dt.date(2022, 1, 1))
        ax.set_yticks([-10, -5, 0, 5, ])
        ax.tick_params(axis = 'x', width = 2.5, length = 5)
        pretty.tickfs(fs, ax = ax)
        ax.tick_params
        fig.savefig(figdir_ew + glkey.replace(' ', '') + 'png', bbox_inches = 'tight', transparent = True)
```

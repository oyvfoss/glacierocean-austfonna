## Glacier velocity time series from itslive 

Downloaded time series of glacier flow velocity from this point:

- 79.47
- 25.02

Here: Reading data and converting units

```python
from oyv.imports import *
import pandas as pd
```

```python
ddir = '/home/oyvindl/work/projects/barents_glaciers/glacierocean-austfonna/data/glacier_vel/'
fn_in = ddir + 'src/itslive-data/LAT79.47--LON25.02.csv'
fn_out = ddir + 'storisstraumen_vel.p'
```

```python
d = pd.read_csv(fn_in)
t_ = date2num(d.mid_date)
sortind = np.argsort(t_)
# T index is jumbled - have to sort
# Checked vs teh web widget that I did do this right
```

```python
D = Bunch()
D.t = np.array(num2date(t_[sortind]))
D.v_m_yr = d['v [m/yr]'][sortind]
D.v_err_m_yr = d['v_error [m/yr]'][sortind]
D.v_m_day = D.v_m_yr/365.25
D.v_err_m_day = D.v_err_m_yr/365.25
```

```python
D.src_script = 'Glacier velocity time series from itslive.ipynb'
D.src_data = fn_in
D.lat = 79.47
D.lon = 25.02
D.meta = '''
 Glacier velocities extracted from NASA ITS LIVE widget for a point on Storisstraumen (%.2f N, %.2f E).
 
 Read from csv and changed to m/day by Ã˜L.
 '''%(D.lat, D.lon)
```

### Save

```python
pickle_save(D, fn_out)
```

### Quick plot

```python
%matplotlib inline
```

```python
fig, ax = plt.subplots(figsize = (10, 4))
ax.fill_between(D.t, (D.v_m_day - D.v_err_m_day).to_numpy(),
                (D.v_m_day + D.v_err_m_day).to_numpy(), color =  col.col('o', sec = True))
ax.plot(D.t, D.v_m_day, '-', color = col.col('b'))
ax.set_ylabel('Glacier flow speed [m/day]')
```
